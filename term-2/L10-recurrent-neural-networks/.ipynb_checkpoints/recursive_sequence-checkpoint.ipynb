The logistic recursive sequence is defined as:
* s<sub>1</sub> = &alpha;
* s<sub>2</sub> = ws<sub>1</sub> - ws<sub>1</sub><sup>2</sup>
* s<sub>3</sub> = ws<sub>2</sub> - ws<sub>1</sub><sup>2</sup>
* s<sub>4</sub> = ws<sub>3</sub> - ws<sub>1</sub><sup>3</sup>

Generate and plot the first 50 elements of this sequence using:
* s<sub>1</sub> = 0.5, w = 1
* s<sub>1</sub> = 0.5, w = 3
* s<sub>1</sub> = 0.0001, w = 4

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

def f_recursive(s1, w, i):
    if i == 0:
        return s1
    s = f_recursive(s1, w, i-1)
    return w*s-w*s**2

inputs = [(0.5, 1), (0.5, 3), (0.0001, 4)]

fig, ax = plt.subplots(len(inputs), 1, figsize=(15, 15))
for i, inp in enumerate(inputs):
    s1, w = inp
    ax[i].plot([f_recursive(s1, w, i) for i in range(50)], '-bo')
    ax[i].set_xlabel('step')
    ax[i].set_ylabel('value')
    ax[i].set_title('s1={}, w={}'.format(s1, w))
fig.tight_layout()
```
