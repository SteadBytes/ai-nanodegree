# Gaussian Blur with OpenCV2
Apply Gaussian filter to reduce noise in an image.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images/noisy_brain.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

## Apply Gaussian Blur

```python
# convert to grayscale to show intensities better for filtering
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# use OpenCV2 GaussianBlur with args (src, kernel_size, standard_deviation)
# bigger filter = bigger blur (use odd dimensions to center pixels in the kernel)
# setstandard_deviatino to 0 for auto calculation
gray_blur = cv2.GaussianBlur(gray, (5,5), 0)

# plot side by side for comparison
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Original Grayscale Image')
ax1.imshow(gray, cmap='gray')

ax2.set_title('Gaussian Blurred Image')
ax2.imshow(gray_blur, cmap='gray')

# zoom in to see reduced speckling
```

## Test Performance with High-Pass Filter
Apply a high-pass filter to original and blurred image to show improved performance in edge detection after removing noise.

```python
sobel_x = np.array([[-2, 0, 2],
                    [-1, 0, 1],
                    [-2, 0, 2]])

# apply high-pass filter to both images
filtered = cv2.filter2D(gray, -1, sobel_x)
filtered_blur = cv2.filter2D(gray_blur, -1, sobel_x)

# plot side by side for comparison
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Original Grayscale Image with High-Pass')
ax1.imshow(filtered, cmap='gray')

ax2.set_title('Gaussian Blurred Image with High-Pass')
ax2.imshow(filtered_blur, cmap='gray')
```

Non-blurred image is much noisier. Blurred image has more accurate edge detection.
