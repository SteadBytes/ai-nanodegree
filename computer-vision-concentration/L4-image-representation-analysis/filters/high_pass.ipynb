# Sobel Vertical Edge Detection Filter

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images/empire_state.jpg')

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

## Convert Image to Grayscale

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

## Create Kernel

```python
# sobel x-direction filter approximates the gradient in the x-direction
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# perform convolution, filter2D takes args (src, ddepth, kernel)
# ddepth = desired depth of output image (-1 means same as input image)
filtered_image = cv2.filter2D(gray, -1, sobel_x)

plt.imshow(filtered_image, cmap='gray')
```

## Create Binary Image
Convert image to purely black and white - useful for making edges stand out.

```python
# sets filtered pixels to white above a certain threshold

retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
```

High-Pass filters can introduce unwanted **noise** by enhancing unwanted features. Usually show up as dots in the filtered image (can see some above).

Using a larger threshold will exaggerate this effect:

```python
retval, binary_image_noisy = cv2.threshold(filtered_image, 20, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image_noisy, cmap='gray')
```
