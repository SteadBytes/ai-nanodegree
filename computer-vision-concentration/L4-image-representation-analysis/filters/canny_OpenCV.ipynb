# Canny Edge Detection in OpenCV

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images/flower.jpg')

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
# grayscale for processing using image intensity
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
# lower and upper thresholds for hysteresis (max intensity range=0-255)
# 1:2 or 1:3 low:high ratios recommended
low = 120
high = 240

edges = cv2.Canny(gray, low, high)

plt.imshow(edges, cmap='gray')
```

```python
# try Canny using different size thresholds
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 180, 240)

f, (ax1, ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title('Wide Threshold')
ax1.imshow(wide, cmap='gray')

ax2.set_title('Tight Threshold')
ax2.imshow(tight, cmap='gray')
```
