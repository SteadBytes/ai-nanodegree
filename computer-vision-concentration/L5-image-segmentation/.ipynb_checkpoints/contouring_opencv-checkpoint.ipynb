# Image Contouring and Contour Features in OpenCV
Aim to detect the thumb pointing **upwards** and create a cropped image of it.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images/thumbs-up-down.jpg')

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

# Create Binary Image

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# binary threshold inverted to show hands as white and background as black
retval, binary = cv2.threshold(gray, 230, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray')
```

## Find and Draw Contours

```python
# findContours takes args src, contour retrieval mode, contour approximation method
# contour approximation method defines which boundary points are stored once detected
# i.e store only some points then fill in the gaps between to approximate

# hierarchy defines the relationship between contours (useful when many contours within one another)
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# draw contours on copy of original image
# drawContours takes args src, contours, contours_to_display, color, size
# -1 for contours_to_display will display all contours
all_contours = cv2.drawContours(np.copy(image_copy), contours, -1, (0,255,0), 2)

plt.imshow(all_contours)
```

## Finding Contour Angles
Can be used to find the **orientation** of an object.

```python
def contour_angles(contours):
    """
    Args:
        contours (list): Contours generated by cv2.findContours
    Returns:
        (list): Angles of contours in same order as input
    """
    angles = []
    for cnt in contours:
        # fit an ellipse around the contour
        # cv2.fitEllipse returns center coordinates, major axis and minor axis lengths, angle
        (x,y), (MA,ma), angle = cv2.fitEllipse(cnt)
        angles.append(angle)
    return angles

angles = contour_angles(contours)
print('Angles of each contour (degrees from x-axis): '+str(angles))
```

## Find Upward Pointing Thumb
Angle/orientation values are degrees from x-axis:
* 0 = flat line
* 90 = perpendicular
Therefore, the 'upwards' thumb will be the contour with the angle value closest to 90

```python
target_angle = 90 # as 'upwards' as possible
upwards_angle_idx = min(range(len(angles)), key=lambda i: abs(angles[i]-target_angle))
# angles and contours lists in same order
upwards_thumb_contour = contours[upwards_angle_idx]
upwards_thumb_highlighted = cv2.drawContours(np.copy(image_copy), [upwards_thumb_contour], -1, (0,255,0), 2)

plt.imshow(upwards_thumb_highlighted)
```

##  Find Bounding Box of Upwards Thumb

```python
x,y,width,height = cv2.boundingRect(upwards_thumb_contour)
upwards_thumb_box = cv2.rectangle(np.copy(image_copy), (x,y), (x+width, y+height), (0,255,0),2)
plt.imshow(upwards_thumb_box)
```

## Crop Image to Upwards Thumb Only

```python
cropped_image = image_copy[y:y+height, x:x+width]

# Display the cropped image side-by-side with the original
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
f.tight_layout()
ax1.imshow(image_copy)
ax1.set_title('Original Image')
ax2.imshow(cropped_image)
ax2.set_title('Cropped Image')
```
