```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

## Image Space to Hough Space

```python
def hough_transform_points(src, theta_step=1):
    """ Transforms src points in image space to rho-theta parameter space
    Args:
        src (array of array): 2D array of points [x,y] coords
        theta_step (int): resolution of theta values (optional)
    Returns:
        tuple of rhos, thetas. Where rhos is 2D array of rho values with length len(src)
    """
    thetas = np.deg2rad(np.arange(-90.0,90.0,theta_step))
    rhos = []
    for x,y in src:
        curr_rhos = []
        for theta in thetas:
            rho = x*np.cos(theta) + y*np.sin(theta)
            curr_rhos.append(rho)
        rhos.append(curr_rhos)
    return rhos, thetas

def hough_transform_binary_img(img, theta_step=1):
    y_idxs, x_idxs = np.nonzero(image)
    thetas = np.deg2rad(np.arange(-90.0,90.0,theta_step))
    rhos = []
    for i in range(len(x_idxs)):
        y = y_idxs[i]
        x = x_idxs[i]
        curr_rhos = []
        for theta in thetas:
            rho = x*np.cos(theta) + y*np.sin(theta)
            curr_rhos.append(rho)
        rhos.append(curr_rhos)
    return rhos, thetas
```

## Detecting Lines with Hough Transform 

```python
def len_rectangle_diag(width, height):
    return np.sqrt((width*width + height*height))

def hough_line(img, theta_step=1, rho_step=1):
    width, height = img.shape
    max_dist = np.ceil(len_rectangle_diag(width, height))
    diag_len=max_dist
    
    rhos = np.linspace(-diag_len, diag_len, diag_len * (2//rho_step))
    thetas = np.deg2rad(np.arange(-90.0,90.0,theta_step))
    
    cos_thetas = np.cos(thetas)
    sin_thetas = np.sin(thetas)
    num_thetas = len(thetas)
    
    accumulator = np.zeros((len(rhos), num_thetas), dtype=np.int64)
    
    y_idxs, x_idxs = np.nonzero(image)
    
    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]
        
        for theta_idx in range(num_thetas):
            
            rho = int(round(x*cos_thetas[theta_idx] + y*sin_thetas[theta_idx]) + diag_len)
            accumulator[rho, theta_idx] +=1
    return accumulator, thetas, rhos
```

```python
image = np.zeros((80,80))
image[10:40, 10:40] = np.eye(30)
accumulator, thetas, rhos = hough_line(image)

# Peak finding based on max votes
rho_idx, theta_idx = np.unravel_index(np.argmax(accumulator), accumulator.shape)
rho = rhos[rho_idx]
theta = thetas[theta_idx]

# rhos, thetas = hough_transform_binary_img(image)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))

ax1.set_title('Input Image')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.imshow(image, cmap='gray')
ax1.invert_yaxis()

ax2.set_title('Hough Transform')
ax2.set_xlabel('Theta')
ax2.set_ylabel('Rho')
ax2.imshow(
        accumulator,cmap='jet',
        extent=[np.rad2deg(thetas[-1]), np.rad2deg(thetas[0]), rhos[-1], rhos[0]])

rho_str= "{0:.2f}".format(rho)
theta_str= "{0:.2f}".format(np.rad2deg(theta))
print("rho={}, theta={}".format(rho_str, theta_str))
print("A line is present in the input image {} pixels from the centre, at an angle of {} degrees.".format(rho_str, theta_str))
```
