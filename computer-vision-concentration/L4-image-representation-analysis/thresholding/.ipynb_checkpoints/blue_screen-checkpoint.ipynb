# Color Thresholding - Blue Screen

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
# read in image and display type and dimensions
image = cv2.imread('images/gandalf.jpg')

print('This image is:', type(image),
     ' with dimensions:', image.shape)

# show image
plt.imshow(image)
```

```python
# opencv uses BGR instead of RGB

# copy BGR image and convert to RGB
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

## Define color threshold

```python
# color selection boundaries in RGB values
# colours that fall within lower_blue and upper_blue range should be very intense blue

# accept no red and green, but a high value for blue
lower_blue = np.array([0, 0, 110])
# accept small amounts of red and green, full blue
upper_blue = np.array([50, 70, 255])
```

## Create a mask

```python
# turns any pixel not in range lower_blue->upper_blue black
mask = cv2.inRange(image_copy, lower_blue, upper_blue)

plt.imshow(mask, cmap='gray')
```

```python
# copy to allow changing mask later
masked_image = np.copy(image_copy)

# mask image to let image show without bluescreen background
# select part of image that overlaps with the part of the mask that is white and replace with black
masked_image[mask != 0] = [0, 0, 0]

plt.imshow(masked_image)
```

## Mask and add a background image

```python
# load and convert background image
background_image = cv2.imread('images/space.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

# crop to correct size
crop_background = background_image[0:554, 0:696]

# mask cropped background so that the 'gandalf' area in blocked
# opposite mask from before, i.e want the image to show through
crop_background[mask==0] = [0, 0, 0]

plt.imshow(crop_background)
```

```python
# Black area is equivalent to zeros in pixel values-> can just add values
complete_image = crop_background + masked_image

plt.imshow(complete_image)
```
